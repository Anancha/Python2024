Any +ve or -ve whole number is called as integer.

````python
int1 = 12
print(int1)
print(type(int1))
````

````python
int2 = -12
print(int2)
print(type(int2))
````

Any +ve or -ve whole number and fractional value together is called as float number.

````python
flt1 = 3.1414
print(flt1)
print(type(flt1))
````

````python
flt2 = -3.1414
print(flt2)
print(type(flt2))
````

Any +ve or -ve number with real and imaginary number is called as complex number

````python
cmplx1 = 12 + 3j
print(cmplx1)
print(type(cmplx1))
````

````python
cmplx2 = -12-3j
print(cmplx2)
print(type(cmplx2))
````

Anything which is either True or False is called as boolean

````python
bool1 = True
print(bool1)
print(type(bool1))
````

````python
bool2 = False
print(bool2)
print(type(bool2))
````

````python
# +ve indexing and +ve slicing

str1 = 'SUBJECTIVISING'
print(str1[0:len(str1):1])
print(str1[3:9:1])
print(str1[10:len(str1):1])
print(str1[0:7:1])
````

````python
# -ve indexing and -ve slicing

print(str1[::-1])
print(str1[-1:-(len(str1)+1):-1])
print(str1[-4:-9:-1])
````

````python
# +ve indexing and -ve slicing

str1[14:0:-1]
````

````python
# -ve indexing and +ve slicing

str1[-(len(str1)+1)::1]
````
